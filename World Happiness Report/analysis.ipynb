```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_csv('dataset/World-Happiness-Report-2023.csv')
df.head(2)
```

```python
fontdict = {'fontsize':16, 'fontweight':'bold'}
title_fontdict = {'fontsize':20, 'fontweight':'bold','fontname':'Times New Roman'}
fig, (ax, ax2) = plt.subplots(2,1,figsize=(14,14),facecolor=(.92, .92, .92))
color = ['#32628d', '#cde01d', '#7eb54e', '#d928ed','#f70a41','#20888a','#f23329','#4f054d','red','#012e1b']

# for ax1 
bars = ax.barh(df['Country name'][:5], df['Ladder score'][:5], color = color)
ax.set_ylabel("Countries",fontdict=fontdict, labelpad=16)
ax.set_xlabel("Score",fontdict=fontdict, labelpad=16)
ax.tick_params(axis='both', which='major', labelsize=13)
ax.bar_label(bars, padding=3, fontsize=12)
ax.set_title("Happiest Countries In The World",fontdict=title_fontdict, pad=20)
ax.invert_yaxis() 

bars_ax2 = ax2.barh(df['Country name'][-5:], df['Ladder score'][-5:], color = color)
ax2.set_ylabel("Countries",fontdict=fontdict, labelpad=16)
ax2.set_xlabel("Score",fontdict=fontdict, labelpad=16)
ax2.tick_params(axis='both', which='major', labelsize=13)
ax2.bar_label(bars_ax2, padding=3, fontsize=12)
ax2.set_title("Least Happy Countries In The World",fontdict=title_fontdict, pad=20)
# ax2.invert_yaxis() 
fig.tight_layout(pad=8.0)


plt.show()
```

```python
def linear_regression(ax,y,x,x_label):

    # Create scatter plot of data points
    ax.scatter(x, y, color='blue')

    # Add labels and title
    ax.set_xlabel(x_label, fontdict=fontdict, labelpad=16)
    ax.set_ylabel('Score', fontdict=fontdict, labelpad=16)
    # ax.set_title('Linear Regression Chart')

    # Add legend
    ax.legend()
```

```python
fontdict = {'fontsize':16, 'fontweight':'bold'}
title_fontdict = {'fontsize':20, 'fontweight':'bold','fontname':'Times New Roman'}
fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(2,3,figsize=(18,14),facecolor=(.92, .92, .92))

linear_regression(ax1,df['Ladder score'],df['Logged GDP per capita'], 'GDP Per Capita')
linear_regression(ax2,df['Ladder score'],df['Social support'], 'Social Support')
linear_regression(ax3,df['Ladder score'],df['Healthy life expectancy'], 'Helathy Life Expectancy')
linear_regression(ax4,df['Ladder score'],df['Freedom to make life choices'], 'Freedom To Make Life Choices')
linear_regression(ax5,df['Ladder score'],df['Generosity'], 'Generosity')
linear_regression(ax6,df['Ladder score'],df['Perceptions of corruption'], 'Perceptions Of Corruption')

fig.suptitle('Correlation Graphs ',fontweight = "bold",fontname="Cursive",fontsize=22,)
plt.show()
```
