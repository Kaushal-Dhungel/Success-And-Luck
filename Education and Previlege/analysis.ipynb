```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
plt.style.use('ggplot')
```

#### 1.Annual Family Income vs Average SAT Scores

```python
df1 = pd.read_csv('dataset/sat-scores-and-income.csv', encoding='utf-16')
df1.head()
```

```python
df1['Total Score'] = df1['Critical Reading'] + df1['Mathematics'] + df1['Writing']
df1.head(3)
```

```python
labels = df1['Income']
width = 0.20  # the width of the bars
x = np.arange(len(labels))  # the label locations
x2 = [i + width for i in x]
x3 = [i + 2*width for i in x]
x4 = [i + 3*width for i in x]


plt.style.use('ggplot')

colors = ["#7eb54e","#32628d","#cde01d", '#20888a']
fig, ax = plt.subplots(figsize=(18,8),facecolor=(.92, .92, .92))
rects1 = ax.bar(x , df1['Critical Reading'], width, label='Critical Reading',color = "#f70a41", edgecolor='black')
rects2 = ax.bar(x2, df1['Mathematics'], width, label='Mathematics', color = "#7eb54e", edgecolor='black')
rects3 = ax.bar(x3, df1['Writing'], width, label='Writing', color = "#20888a", edgecolor='black')
rects4 = ax.bar(x4, df1['Total Score'], width, label='Total', color = "#32628d", edgecolor='black')



# Add some text for labels, title and custom x-axis tick labels, etc.
fontdict = {'fontsize':18,}
ax.set_xlabel('Annual Family Income',labelpad=20, fontdict=fontdict )
ax.set_ylabel('Average SAT Scores',labelpad=20, fontdict=fontdict)
ax.set_title('Annual Family Income vs Average SAT Scores')
ax.set_xticks(x2, labels,weight = 'bold',rotation=10)
ax.tick_params(axis='y', which='major', labelsize=12)
ax.legend(loc='upper left')

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)


fig.tight_layout()

plt.show()
```

### 2. Parental Education vs SAT Scores

```python
df2 = pd.read_csv('dataset/sat-scores-and-parental-education.csv', encoding='utf-16')
df2.head(3)
```

```python
df2['Total Score'] = df2['Critical Reading'] + df2['Mathematics'] + df2['Writing']
df2.head(3)
```

```python
labels = df2['Highest Level of Parental Education']
width = 0.20  # the width of the bars
x = np.arange(len(labels))  # the label locations
x2 = [i + width for i in x]
x3 = [i + 2*width for i in x]
x4 = [i + 3*width for i in x]

plt.style.use('ggplot')

colors = ["#7eb54e","#32628d","#cde01d"]
fig, ax = plt.subplots(figsize=(14,8),facecolor=(.92, .92, .92))
rects1 = ax.bar(x , df2['Critical Reading'], width, label='Critical Reading',color = "#f70a41", edgecolor='black')
rects2 = ax.bar(x2, df2['Mathematics'], width, label='Mathematics', color = "#7eb54e", edgecolor='black')
rects3 = ax.bar(x3, df2['Writing'], width, label='Writing', color = "#20888a", edgecolor='black')
rects4 = ax.bar(x4, df2['Total Score'], width, label='Total', color = "#32628d", edgecolor='black')


# Add some text for labels, title and custom x-axis tick labels, etc.
fontdict = {'fontsize':18,}
ax.set_xlabel('Parental Education',labelpad=20, fontdict=fontdict )
ax.set_ylabel('Average SAT Scores',labelpad=20, fontdict=fontdict)
ax.set_title('Parental Education vs Average SAT Scores')
ax.set_xticks(x2, labels,weight = 'bold')
ax.tick_params(axis='y', which='major', labelsize=12)
ax.legend(loc='upper left')

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

fig.tight_layout()

plt.show()
```
